```python
import matplotlib.pyplot as plt
```

```python
savings = [0, 150, 500, 1000, 1500, 2000, 3000, 4000, 5000, 7000, 8500, 10000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
plt.plot(months, savings)
plt.xlabel('Months')
plt.ylabel('Savings ($)')
plt.title('My First Plot')
plt.show()
```
